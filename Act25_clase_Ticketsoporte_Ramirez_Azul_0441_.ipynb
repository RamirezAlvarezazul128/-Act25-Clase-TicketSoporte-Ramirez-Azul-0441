## Clase Ticket Soporte
Este programa utiliza la Programación Orientada a Objetos para digitalizar el control de reportes técnicos.
![link text](https://raw.githubusercontent.com/RamirezAlvarezazul128/-Act25-Clase-TicketSoporte-Ramirez-Azul-0441/refs/heads/main/info%20clase%20Ticket%20soporte%20Act%2025.png)

```python
# Lenguaje: Python - Nivel: Principiante
# Clase: TicketSoporte

class TicketSoporte:
    # 1. Atributos inicializados mediante el constructor
    def __init__(self, folio, problema, estado):
        self.folio = folio
        self.problema = problema
        self.estado = estado

    # 2. Función: mostrarTicket() - Muestra la información básica
    def mostrarTicket(self):
        print(f"🎫 TICKET #{self.folio}: Procesando reporte técnico...")

    # 3. Función: verificarEstado() - Indica en qué fase se encuentra
    def verificarEstado(self):
        print(f"🔍 ESTADO: El ticket se encuentra actualmente {self.estado}.")

    # 4. Función: archivarProblema() - Simula el cierre o guardado
    def archivarProblema(self):
        print(f"📁 ARCHIVADO: La descripción '{self.problema}' ha sido guardada en la base de datos.")

# Función para interactuar con el usuario y pedir datos (Igual al ejemplo de SalaCine)
def pedir_datos(nombre_objeto):
    print(f"\n--- Ingrese los atributos para {nombre_objeto} ---")
    fol = input("Ingrese el número de Folio: ")
    prob = input("Describa el problema técnico: ")
    est = input("Ingrese el estado (Pendiente/En Proceso/Resuelto): ")
    return TicketSoporte(fol, prob, est)

# --- Creación de los 2 objetos solicitados ---
ticket1 = pedir_datos("ticket1")
ticket2 = pedir_datos("ticket2")

# --- Ejecución de las actividades (Funciones agrupadas por objeto) ---
print("\n" + "="*45)
print("REPORTE DE TICKETS DE SOPORTE")
print("="*45)

# Actividades para el objeto ticket1 (Todas juntas)
print(f"\n>>> GESTIONANDO: Folio {ticket1.folio}")
ticket1.mostrarTicket()
ticket1.verificarEstado()
ticket1.archivarProblema()

print("-" * 30) # Separador visual

# Actividades para el objeto ticket2 (Todas juntas)
print(f"\n>>> GESTIONANDO: Folio {ticket2.folio}")
ticket2.mostrarTicket()
ticket2.verificarEstado()
ticket2.archivarProblema()

print("\n" + "="*45)
print("fin del programa")
print("programa realizado por Ramirez Azul NL = 44")
```
